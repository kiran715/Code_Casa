IMPORTING THE LIBRARIES AND LOADING THE DATASET

```python
import pandas as pd
import numpy as np
df = pd.read_csv("creditcard.csv")
df.head()
```

```python
df.info()
```

STATISTICAL ANALYSIS

```python
df.describe().T
```

```python
df.isna().sum()
```

REMOVING OUTLIERS USING QUANTILE METHOD

```python
def remove_outliers(col):
    min_thresold,max_thresold = df[col].quantile([0.01,0.999]) 
    return min_thresold,max_thresold
for col_name in df.iloc[:,1:29].columns:
    lower_limit,upper_limit = remove_outliers(col_name)
    df[col_name] = np.where(df[col_name] > upper_limit,upper_limit,np.where(df[col_name] < lower_limit,lower_limit,df[col_name]))
    
```

```python
df.describe().T
```

```python
df["Class"].value_counts()
```

```python
non_fraud = df[df["Class"]==0] #0 -> non fradulent
fraud = df[df["Class"]==1] # 1 -> fradulent 
```

```python
non_fraud.Amount.describe().T
```

```python
fraud.Amount.describe().T
```

UNDERSMAPLING THE DATA SINCE THE DATA WAS UNBALANCED

```python
Non_fraud = non_fraud.sample(n=492)
data = pd.concat([Non_fraud,fraud],axis=0)
```

```python
data["Class"].value_counts()
```

```python
x = data.drop(["Class"],axis=1)
y = data["Class"]
x.shape,y.shape
```

SPLITTING THE DATA FOR TRAIN AND TEST

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.2,random_state=2,stratify=y)
print(X_train.shape,X_test.shape,Y_train.shape,Y_test.shape)
```

```python
MODEL SELECTRION AND TRAINING 
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

CHECKING ACCURACY ON TRAINING DATA

```python
from sklearn.metrics import accuracy_score,confusion_matrix
X_train_predict = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_predict, Y_train)
print("Accuracy on train data : ",training_data_accuracy)
```

```python
CHECKING ACCURACY ON TEST DATA
```

```python
X_test_predict = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_predict, Y_test)
print("Accuracy on test data : ",testing_data_accuracy)
```

VISUALIZING THE CONFUSION MATRIX

```python
import seaborn as sns
import matplotlib.pyplot as plt
cm_train = confusion_matrix(X_train_predict, Y_train)
sns.heatmap(cm_train,annot=True,fmt='g',xticklabels=['Not fraud','fraud'],yticklabels=['Not fraud','fraud'])
plt.xlabel('Actual',fontsize=12)
plt.ylabel('Prediction',fontsize=12)
plt.title('Confusion Matrix for Train data',fontsize=20)
plt.show()
```

```python
cm_test = confusion_matrix(X_test_predict, Y_test)
sns.heatmap(cm_test,annot=True,fmt='g',xticklabels=['Not fraud','fraud'],yticklabels=['Not fraud','fraud'])
plt.xlabel('Actual',fontsize=12)
plt.ylabel('Prediction',fontsize=12)
plt.title('Confusion Matrix for Test data',fontsize=20)
plt.show()
```
